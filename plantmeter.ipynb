```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# set figure size
plt.rcParams['figure.figsize'] = [16, 9]
```

```python
file_name = './010_2_08262019_P5_NFACC39-1.5_L.jpg'
im = cv2.imread(file_name)
im[:,:,0]
```

```python

fig, axs = plt.subplots(1,3)
axs[0].imshow(im[:,:,0])
axs[1].imshow(im[:,:,1])
axs[2].imshow(im[:,:,2])
plt.show()
```

```python
grey = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
grey.shape
plt.imshow(grey[:,:])
```

```python
grey_denoised = cv2.medianBlur(grey, 15)
grey_blur = cv2.medianBlur(grey, 151)
plt.imshow(grey_denoised)
grey_diff = cv2.subtract(grey_blur, grey_denoised)
_, binary = cv2.threshold(grey_diff, 15, 255, cv2.THRESH_BINARY)
prop_plant = np.count_nonzero(binary) / (binary.shape[0] *  binary.shape[1])
prop_plant
```
